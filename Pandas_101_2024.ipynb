-วนลูปนับว่านิทานที่มีคำว่าแม่กับพ่อในชื่อเรื่องทั้งหมดกี่เรื่อง เรื่องอะไรบ้าง
-มีคำดังกล่าาวในนิทานชีวิตหรือนิทานประจำถิ่นมากกว่ากัน
-คนแต่งคนไหนชอบแต่งนิทานที่มีชื่อเรื่องว่าแม่หรือพ่อมากที่สุด

1.วนลูปนับว่านิทานที่มีคำว่าแม่กับพ่อในชื่อเรื่องทั้งหมดกี่เรื่อง เรื่องอะไรบ้าง

````
count_mom = 0
count_dad = 0
mom_story = []
dad_story = []
for each_row in folk_df.iterrows():
    if 'แม่' in each_row[1]['title']:
        count_mom += 1
        mom_story.append(each_row[1]['title'])
    if 'พ่อ' in each_row[1]['title']:
        count_dad += 1
        dad_story.append(each_row[1]['title'])

print(f'นิทานที่มีคำว่าแม่ในชื่อเรื่องมี {count_mom} เรื่อง')
print(f'เรื่องดังนี้ \n {mom_story}')
print(f'นิทานที่มีคำว่าพ่อในชื่อเรื่องมี {count_dad} เรื่อง')
print(f'เรื่องดังนี้ \n {dad_story}')
````

2.มีคำดังกล่าาวในนิทานชีวิตหรือนิทานประจำถิ่นมากกว่ากัน

````
count_mom_life = 0
count_dad_life = 0
count_mom_local = 0
count_dad_local = 0

for each_row in folk_df.iterrows():
    if 'แม่' in each_row[1]['title']:
        if each_row[1]['type'] == 'นิทานชีวิต':
            count_mom_life += 1
        elif each_row[1]['type'] == 'นิทานประจำถิ่น':
            count_mom_local += 1
    if 'พ่อ' in each_row[1]['title']:
        if each_row[1]['type'] == 'นิทานชีวิต':
            count_dad_life += 1
        elif each_row[1]['type'] == 'นิทานประจำถิ่น':
            count_dad_local += 1

print(f'นิทานชีวิตที่มีคำว่าแม่ในชื่อเรื่องมี {count_mom_life} เรื่อง')
print(f'นิทานประจำถิ่นที่มีคำว่าแม่ในชื่อเรื่องมี {count_mom_local} เรื่อง')
print(f'นิทานชีวิตที่มีคำว่าพ่อในชื่อเรื่องมี {count_dad_life} เรื่อง')
print(f'นิทานประจำถิ่นที่มีคำว่าพ่อในชื่อเรื่องมี {count_dad_local} เรื่อง')

if (count_mom_life + count_dad_life) > (count_mom_local + count_dad_local):
    print('มีคำว่าพ่อหรือแม่ในนิทานชีวิตมากกว่า')
elif (count_mom_life + count_dad_life) < (count_mom_local + count_dad_local):
    print('มีคำว่าพ่อหรือแม่ในนิทานประจำถิ่นมากกว่า')
else:
    print('มีคำว่าพ่อหรือแม่ในนิทานชีวิตและนิทานประจำถิ่นเท่ากัน')
````

3.คนแต่งคนไหนชอบแต่งนิทานที่มีชื่อเรื่องว่าแม่หรือพ่อมากที่สุด

````
creator_counts = {}
for each_row in folk_df.iterrows():
    creator = each_row[1]['creator']
    if 'แม่' in each_row[1]['title'] or 'พ่อ' in each_row[1]['title']:
        if creator not in creator_counts:
            creator_counts[creator] = 0
        creator_counts[creator] += 1

max_count = 0
most_frequent_creator = ''
for creator, count in creator_counts.items():
    if count > max_count:
        max_count = count
        most_frequent_creator = creator

print(f"คนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่าแม่หรือพ่อมากที่สุดคือ: {most_frequent_creator} ({max_count} เรื่อง)")
````

บรรยายเกี่ยวกับผู้กำกับชั้นนำของอินเดีย 3 คน ว่าทำไมเราถึงคิดว่า เป็นผู้กำกับชั้นนำ และชั้นนำยังไง

````
import pandas as pd
movie_df = pd.read_csv('/content/IMDb Movies India.csv', encoding='latin-1')
print(movie_df.shape)
movie_df.head()
````

````
# prompt: name ที่มี rating มากที่สุด 3 อันดับ เป็นของ director คนใดบ้าง

# หา 3 อันดับแรกของ rating
top_3_ratings = movie_df.nlargest(3, 'Rating')

# แสดงชื่อ director ของ 3 อันดับแรก
for index, row in top_3_ratings.iterrows():
    print(f"Director of '{row['Name']}' (Rating: {row['Rating']}): {row['Director']}")
````
