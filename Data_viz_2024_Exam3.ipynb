#จุดที่น่าสนใจของข้อมูลจุดที่1
คอมเมนต์ที่มียอดไลก์มากที่สุดแสดงให้เห็นว่าคอมเมนต์นั้นอาจมีความคิดเห็นที่ตรงกับความรู้สึกของหลาย ๆ คน อาจพูดถึงสิ่งที่เกี่ยวข้องกับประสบการณ์ส่วนตัวของตัวเองหรือข้อมูลที่มีประโยชน์หรือมีข้อเสนอแนะที่ดี
โดยคอมเมนต์ที่เป็น Top1 เล่าประสบการณ์เกี่ยวกับมะเร็งปากมดลูก

```python
# prompt: ชนิดของมะเร็งที่อาจเกี่ยวข้อง

# สร้าง list ของคำที่เกี่ยวข้องกับชนิดของมะเร็ง
cancer_keywords = ["เนื้องอก", "มะเร็งเต้านม", "มะเร็งปอด", "มะเร็งลำไส้", "มะเร็งตับ",
                   "มะเร็งต่อมลูกหมาก", "มะเร็งปากมดลูก", "มะเร็งผิวหนัง", "มะเร็งกระเพาะอาหาร",
                   "มะเร็งเม็ดเลือดขาว", "มะเร็งต่อมน้ำเหลือง", "มะเร็งสมอง", "มะเร็งกระดูก", "มะเร็งรังไข่",
                   "มะเร็งอัณฑะ", "มะเร็งตับอ่อน", "มะเร็งไต", "มะเร็งกระเพาะปัสสาวะ", "มะเร็งหลอดอาหาร",
                   "มะเร็งช่องปาก", "มะเร็งคอหอย", "มะเร็งกล่องเสียง", "มะเร็งต่อมไทรอยด์"]
cancer_counts = {}
for index, row in docter.iterrows():
  detail = row['comments']
  for cancer in cancer_keywords:
    if cancer in detail:
      if cancer in cancer_counts:
        cancer_counts[cancer] += 1
      else:
        cancer_counts[cancer] = 1
#เรียงลำดับจากน้อยไปมาก
sorted_cancer_deaths = sorted(cancer_counts.items(), key=lambda item: item[1], reverse=True)
sorted_cancer_deaths
```
