## IMDb Movies India table

```python
# prompt: open csv file name IMDb Movies India.csv and set csv variable name is  movies_india
import pandas as pd
movies_india = pd.read_csv("IMDb Movies India.csv", encoding='latin-1') # Try reading the file with 'latin-1' encoding.
print(movies_india.shape)
movies_india.head()
```

```python
# prompt: Using dataframe movies_india:  เช็กค่า missing  ตาราง

# Check for missing values in the dataframe
missing_values = movies_india.isnull().sum()

# Display the count of missing values for each column
print(missing_values)
```

### Column: "Name"

```python
# prompt: movies_india: แทนที่ แถวแรก ของ column Name ด้วย Unknown
movies_india.loc[0, 'Name'] = 'Unknown'
movies_india
```

```python
# prompt: movies_india: explore column Name

# Explore the 'Name' column in movies_india DataFrame

# Check the number of unique movie names
unique_movie_names = movies_india['Name'].nunique()
print(f"Number of unique movie names: {unique_movie_names}")

# Check the most frequent movie names
most_frequent_movie_names = movies_india['Name'].value_counts().head(10)
print("\nMost frequent movie names:\n", most_frequent_movie_names)

# Check for missing values in the 'Name' column
missing_movie_names = movies_india['Name'].isnull().sum()
print(f"\nNumber of missing movie names: {missing_movie_names}")
```

### Column: "Year"

```python
# prompt: Using dataframe movies_india: explore column Year

# Look at the unique values in the 'Year' column
print(movies_india['Year'].unique())

# Count the number of movies released in each year
year_counts = movies_india['Year'].value_counts()
print(year_counts)

# Calculate some descriptive statistics for the 'Year' column
print(movies_india['Year'].describe())
```

```python
# จัดการ column Year จาก '(2004)' ให้เป็นข้อมูล  '2004'

import pandas as pd
movies_india['Year'] = movies_india['Year'].str.extract('(\d{4})')
# movies_india['Year'] = pd.to_numeric(movies_india['Year'], errors='coerce')

# Look at the unique values in the 'Year' column
print(movies_india['Year'].unique())
```

```python
# prompt: movies_india: เติม missing ของ column Year ด้วย "Unknown"

movies_india['Year'] = movies_india['Year'].fillna('Unknown')

# Look at the unique values in the 'Year' column
print(movies_india['Year'].unique())
```

### Column: "Duration"

```python
# prompt: movies_india: explore column Duration

import pandas as pd
import matplotlib.pyplot as plt
# Explore the 'Duration' column
print(movies_india['Duration'].unique())

# Check for missing values
print('Check for missing values ', movies_india['Duration'].isnull().sum())
```

```python
# จัดการ column Duration จาก '235 min'  ให้เป็นข้อมูล 235

# Remove ' min' from the 'Duration' column and convert to numeric
movies_india['Duration'] = movies_india['Duration'].str.replace(' min', '').fillna(0).astype(int)

# Display the updated 'Duration' column
print(movies_india['Duration'].head())
```

```python
# เติม 0 ของ column Duration ด้วย "Unknown" เพราะเป็นค่า missing

# Replace 0 values in 'Duration' with 'Unknown'
movies_india['Duration'] = movies_india['Duration'].replace(0, 'Unknown')

# Display the updated 'Duration' column
print(movies_india['Duration'].head())
```

### Column: "Genre"

```python
# prompt: Using dataframe movies_india: explore column Genre

# Look at the unique genres
print(movies_india['Genre'].unique())

# Count the number of movies in each genre
genre_counts = movies_india.groupby('Genre')['Name'].count()
print(genre_counts)

# Find the most frequent genres
print(genre_counts.sort_values(ascending=False).head(10))
```

```python
# prompt: movies_india: split value in column genre by ',' and make more column for each of those

# Split the 'Genre' column by ',' and create new columns
movies_india[['Genre1', 'Genre2', 'Genre3']] = movies_india['Genre'].str.split(',', expand=True).fillna('Unknown')

# Display the updated dataframe
movies_india.head()
```

```python
# prompt: Using dataframe movies_india: เเทนค่า missing ของคอลัมน์ Rating,	Votes,	Director, Actor 1, Actor 2, Actor 3 ด้วย Unknown

# Replace missing values in specified columns with "Unknown"
columns_to_fill = ['Rating', 'Votes', 'Director', 'Actor 1', 'Actor 2', 'Actor 3']
for column in columns_to_fill:
  movies_india[column] = movies_india[column].fillna('Unknown')
movies_india
```

```python
# prompt: Using dataframe movies_india:  เช็กค่า missing  ตาราง

# Check for missing values in the dataframe
missing_values = movies_india.isnull().sum()

# Display the count of missing values for each column
print(missing_values)
```

```python
import numpy as np
import pandas as pd  #import pandas มาเพราะข้อมูลจะเกี่ยวกับตาราง
import matplotlib.pyplot as plt
from math import pi #import math มาเนื่องจากเกี่ยวกับวงกลมนำมาใช้เพื่อดึงค่าพาย
```

```python
# prompt: Using dataframe top สร้าง Bar 10อันดับประเภทหนังRatingที่สูง2017-2021

# Filter the DataFrame for movies released between 2017 and 2021
movies_2017_2021 = movies_india[(movies_india['Year'] >= '2017') & (movies_india['Year'] <= '2021')]

# Convert 'Rating' column to numeric, handling potential errors
movies_2017_2021['Rating'] = pd.to_numeric(movies_2017_2021['Rating'], errors='coerce')

# Group by genre and calculate the average rating
genre_ratings = movies_2017_2021.groupby('Genre')['Rating'].mean()

# Sort the average ratings in descending order and get the top 10 genres
top_10_genres = genre_ratings.sort_values(ascending=False).head(10)

# Create a bar plot of the top 10 genres with highest average rating
plt.figure(figsize=(12, 6))
plt.bar(top_10_genres.index, top_10_genres.values)
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.title('Top 10 Genres with Highest Average Rating (2017-2021)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: Using dataframe top สร้าง Bar 10อันดับประเภทหนังที่มีเยอะ 2017-2021

# Filter the DataFrame for movies released between 2017 and 2021
movies_2017_2021 = movies_india[(movies_india['Year'] >= '2017') & (movies_india['Year'] <= '2021')]

# Group by genre and count the number of movies
genre_counts_2017_2021 = movies_2017_2021.groupby('Genre')['Name'].count()

# Sort the counts in descending order and get the top 10 genres
top_10_genres_2017_2021 = genre_counts_2017_2021.sort_values(ascending=False).head(10)

# Create a bar plot of the top 10 genres
plt.figure(figsize=(12, 6))
plt.bar(top_10_genres_2017_2021.index, top_10_genres_2017_2021.values)
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.title('Top 10 Genres with Most Movies (2017-2021)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: สร้างrader char เปรียบเทียบGenreในแต่ละปี10อันดับตั้งแต่ปี2012-2021

import matplotlib.pyplot as plt
import numpy as np

# Filter the DataFrame for movies released between 2017 and 2021
movies_2017_2021 = movies_india[(movies_india['Year'] >= '2017') & (movies_india['Year'] <= '2021')]

# Group by year and genre, then count the number of movies
genre_counts_by_year = movies_2017_2021.groupby(['Year', 'Genre'])['Name'].count().unstack().fillna(0)

# Get the top 10 genres for each year
top_10_genres_by_year = {}
for year in genre_counts_by_year.index:
  top_10_genres_by_year[year] = genre_counts_by_year.loc[year].nlargest(10).index.tolist()

# Create a list of unique genres across all years
all_genres = set()
for year in top_10_genres_by_year:
  all_genres.update(top_10_genres_by_year[year])

# Create a radar chart for each year
labels = list(all_genres)
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

for year in genre_counts_by_year.index:
  values = genre_counts_by_year.loc[year][list(all_genres)].fillna(0).tolist()
  values += values[:1]
  ax.plot(angles, values, linewidth=1, label=year)
  ax.fill(angles, values, alpha=0.25)


ax.set_thetagrids(np.degrees(angles[:-1]), labels)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.title('Top 10 Genres Comparison by Year (2017-2021)')

plt.show()
```

```python
# prompt: สร้างrader char เปรียบเทียบRatingในแต่ละปี5อันดับ-ของหนังแต่ละประเภทตั้งแต่ปี2017-2021

# Filter the DataFrame for movies released between 2017 and 2021 and Rating not equal to 'Unknown'
movies_2017_2021_rating = movies_india[(movies_india['Year'] >= '2017') & (movies_india['Year'] <= '2021') & (movies_india['Rating'] != 'Unknown')]

# Convert Rating column to numeric for sorting and plotting
movies_2017_2021_rating['Rating'] = pd.to_numeric(movies_2017_2021_rating['Rating'], errors='coerce')

# Group by year and genre, then calculate the average rating
genre_rating_by_year = movies_2017_2021_rating.groupby(['Year', 'Genre'])['Rating'].mean().unstack().fillna(0)

# Get the top 5 genres with the highest average rating for each year
top_5_genres_by_year = {}
for year in genre_rating_by_year.index:
  top_5_genres_by_year[year] = genre_rating_by_year.loc[year].nlargest(5).index.tolist()

# Create a list of unique genres across all years
all_genres = set()
for year in top_5_genres_by_year:
  all_genres.update(top_5_genres_by_year[year])

# Create a radar chart for each year
labels = list(all_genres)
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

for year in genre_rating_by_year.index:
  values = genre_rating_by_year.loc[year][list(all_genres)].fillna(0).tolist()
  values += values[:1]
  ax.plot(angles, values, linewidth=1, label=year)
  ax.fill(angles, values, alpha=0.25)


ax.set_thetagrids(np.degrees(angles[:-1]), labels)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.title('Top 5 Genres Average Rating Comparison by Year (2017-2021)')

plt.show()
```
